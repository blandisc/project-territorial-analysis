```python
import pandas as pd
import geopandas as gp
```

# Import data with population characteristics

```python
censo = pd.read_csv("../static/Data/Poblacion/conjunto_de_datos/resultados_ageb_urbana_09_cpv2010.csv")
```

```python
poblacionales = censo[censo.mza>0]
poblacionales.head()
```

```python
poblacionales = poblacionales[(poblacionales.nom_mun == "Cuajimalpa de Morelos")|(poblacionales.nom_mun == "Álvaro Obregón")|(poblacionales.nom_mun == "Benito Juárez")]
poblacionales
```

```python
poblacionales_df = poblacionales[['entidad','nom_ent','mun','nom_mun','loc','nom_loc','ageb','mza','tothog','pobtot','prom_hnv','p15ym_se','p18ym_pb','graproes','pea','pder_ss','vph_3ymasc','vph_excsa','vph_c_serv','vph_inter']]
poblacionales_df.head(10)
```

```python
poblacionales_df['tothog'] = pd.to_numeric(poblacionales_df['tothog'], errors='coerce').fillna(0)
poblacionales_df['prom_hnv'] = pd.to_numeric(poblacionales_df['prom_hnv'], errors='coerce').fillna(0)
poblacionales_df['p15ym_se'] = pd.to_numeric(poblacionales_df['p15ym_se'], errors='coerce').fillna(0)
poblacionales_df['p18ym_pb'] = pd.to_numeric(poblacionales_df['p18ym_pb'], errors='coerce').fillna(0)
poblacionales_df['graproes'] = pd.to_numeric(poblacionales_df['graproes'], errors='coerce').fillna(0)
poblacionales_df['pea'] = pd.to_numeric(poblacionales_df['pea'], errors='coerce').fillna(0)
poblacionales_df['pder_ss'] = pd.to_numeric(poblacionales_df['pder_ss'], errors='coerce').fillna(0)
poblacionales_df['vph_3ymasc'] = pd.to_numeric(poblacionales_df['vph_3ymasc'], errors='coerce').fillna(0)
poblacionales_df['vph_excsa'] = pd.to_numeric(poblacionales_df['vph_excsa'], errors='coerce').fillna(0)
poblacionales_df['vph_c_serv'] = pd.to_numeric(poblacionales_df['vph_c_serv'], errors='coerce').fillna(0)
poblacionales_df['vph_inter'] = pd.to_numeric(poblacionales_df['vph_inter'], errors='coerce').fillna(0)
poblacionales_df.dtypes
```

## Scorecard for measuring wealth

```python
poblacionales_df["prom_hnv_s"] = poblacionales_df['prom_hnv'].apply(lambda x:1 if x <= 1.5 else 0)
poblacionales_df["p15ym_se_s"] = (poblacionales_df['p15ym_se']/poblacionales_df['pobtot']).apply(lambda x:1 if x < 0.01 else 0)
poblacionales_df["p18ym_pb_s"] = (poblacionales_df['p18ym_pb']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.40 else 0)
poblacionales_df["graproes_s"] = poblacionales_df['graproes'].apply(lambda x:1 if x >= 10 else 0)
poblacionales_df["pea_s"] = (poblacionales_df['pea']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.40 else 0)
poblacionales_df["pder_ss_s"] = (poblacionales_df['pder_ss']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.60 else 0)
poblacionales_df["vph_3ymasc_s"] = (poblacionales_df['vph_3ymasc']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.20 else 0)
poblacionales_df["vph_excsa_s"] = (poblacionales_df['vph_excsa']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.25 else 0)
poblacionales_df["vph_c_serv_s"] = (poblacionales_df['vph_c_serv']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.25 else 0)
poblacionales_df["vph_inter_s"] = (poblacionales_df['vph_inter']/poblacionales_df['pobtot']).apply(lambda x:1 if x > 0.12 else 0)

poblacionales_df
```

```python
poblacionales_df.dtypes
```

```python
poblacionales_df["score"] = poblacionales_df.iloc[:,20:30].sum(axis=1)
poblacionales_df
```

```python
poblacionales_df["CVE_ENT"] = poblacionales_df.entidad.astype(str).str.zfill(2)
poblacionales_df["CVE_MUN"] = poblacionales_df.mun.astype(str).str.zfill(3)
poblacionales_df["CVE_LOC"] = poblacionales_df["loc"].astype(str).str.zfill(4)
poblacionales_df["CVE_AGEB"] = poblacionales_df.ageb.astype(str).str.zfill(4)
poblacionales_df["CVE_MZA"] = poblacionales_df.mza.astype(str).str.zfill(3)
```

```python
poblacionales_df
```

## Add Key

```python
poblacionales_df["KEY"] = poblacionales_df.apply(lambda row: row.CVE_ENT + row.CVE_MUN + row.CVE_LOC + row.CVE_AGEB + row.CVE_MZA, axis=1)
```

```python
poblacionales_df
```

```python
poblacionales_df = poblacionales_df.rename(columns={"KEY":"CVEGEO"})
```

```python
poblacionales_df
```

```python
poblacionales_df = poblacionales_df.set_index('CVEGEO')
poblacionales_df.head(10)
```

# Import shp file

```python
poligonos = gp.read_file("../static/Data/Archivos SHP/conjunto_de_datos/09m.shp")
```

```python
poligonos
```

```python
poligonos = poligonos[(poligonos.CVE_MUN == "004")|(poligonos.CVE_MUN == "010")|(poligonos.CVE_MUN == "014")]
poligonos
```

```python
poligonos = poligonos.set_index('CVEGEO')
poligonos
```

```python
master_df = poligonos.merge(poblacionales_df, on=["CVEGEO"], how="left")
```

```python
master_df
```

# Convert to geoJSON

```python
master_geojson = master_df.to_crs("EPSG:4326")
master_geojson.head()
```

```python
master_geojson["geometry"] = master_geojson["geometry"].simplify(tolerance = 0.0001, preserve_topology = True)
```
