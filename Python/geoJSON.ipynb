```python
import pandas as pd
import geopandas as gp
```

```python
censo = pd.read_csv("./Características de la Población/conjunto_de_datos/resultados_ageb_urbana_09_cpv2010.csv")
```

```python
poblacionales = censo[censo.mza>0]
poblacionales.head()
```

```python
poblacionales_df = poblacionales[['entidad','nom_ent','mun','nom_mun','loc','nom_loc','ageb','mza','pobtot','pocupada','pdesocup','pea','pea_m','pea_f','pe_inac','pe_inac_m','pe_inac_f']]
poblacionales_df.head(10)
```

```python
poblacionales_df["CVE_ENT"] = poblacionales_df.entidad.astype(str).str.zfill(2)
poblacionales_df["CVE_MUN"] = poblacionales_df.mun.astype(str).str.zfill(3)
poblacionales_df["CVE_LOC"] = poblacionales_df["loc"].astype(str).str.zfill(4)
poblacionales_df["CVE_AGEB"] = poblacionales_df.ageb.astype(str).str.zfill(4)
poblacionales_df["CVE_MZA"] = poblacionales_df.mza.astype(str).str.zfill(3)
```

```python
poblacionales_df
```

```python
poblacionales_df["KEY"] = poblacionales_df.apply(lambda row: row.CVE_ENT + row.CVE_MUN + row.CVE_LOC + row.CVE_AGEB + row.CVE_MZA, axis=1)
```

```python
poblacionales_df
```

```python
poblacionales_df = poblacionales_df.rename(columns={"KEY":"CVEGEO"})
```

```python
poblacionales_df
```

```python
poblacionales_df = poblacionales_df[['CVEGEO','entidad','nom_ent','mun','nom_mun','loc','nom_loc','ageb','mza','pobtot','pocupada','pdesocup','pea','pea_m','pea_f','pe_inac','pe_inac_m','pe_inac_f']]
poblacionales_df.head(10)
```

```python
poligonos = gp.read_file("./Archivos SHP/conjunto_de_datos/09m.shp")
```

```python
poligonos
```

```python
master_df = poligonos.merge(poblacionales_df, on=["CVEGEO"], how="left")
```

```python
master_df
```

```python
master_df.to_csv("./master_df.csv")
```

```python
master_geojson = master_df.to_crs("EPSG:4326")
master_geojson.head()
```

```python
master_geojson["geometry"] = master_geojson["geometry"].simplify(tolerance = 0.001,preserve_topology = True)
```

```python
master_geojson.to_file("master_geojson.geojson",driver="GeoJSON")
```
